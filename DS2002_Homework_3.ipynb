````python
import pandas as pd
````

````python
df = pd.read_csv('/content/acc_players-2324.csv', sep=',')
````

````python
df.head()
````

````python
df.describe()
````

````python
df.dropna()
````

````python
print(df['PTS'].sum(skipna = True))
````

````python
print(df.loc[df['MP'] == df['MP'].max(), 'Player'])
````

````python
print(df[['Player', 'TRB']].sort_values(by = 'TRB', ascending = False).head(5))
````

````python
new_df = df.loc[df['MP'] >= 500]
````

````python
new_df.head()
````

````python
print(new_df[['Player', 'AST']].sort_values(by = 'AST', ascending = False).head(1))
````

````python
print(new_df[['Player', 'AST']].sort_values(by = 'AST', ascending = False).head(3))
````

````python
print(new_df[['Player', 'BLK']].sort_values(by = 'BLK', ascending = False).head(3))
````

````python
print(df.groupby('School')['PTS'].sum())
````

````python
print(df.groupby('School')['AST'].sum())
````

````python
print(df.groupby('School')['PTS'].sum().sort_values(ascending = False).head(3))
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
sns.set(style="whitegrid")
````

````python
bc_df = df.sort_values(by = 'PTS', ascending = False).head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x = 'Player', y = 'PTS', data = bc_df)
plt.title('Top 5 Point Scorers')
plt.xlabel('Player')
plt.ylabel('Total Points Scored')
plt.show()
````

````python
plt.figure(figsize=(10, 6))
plt.scatter(df['FG%'], df['WS'])
plt.title('Field Goal Percentage v. Win Shares')
plt.xlabel('Field Goal Percentage')
plt.ylabel('Win Shares')
plt.show()
````

````python
cc = df['FG%'].corr(df['WS'])
print(cc)
````
