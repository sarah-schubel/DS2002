```python
import requests
import pandas as pd
import json
import os
import sqlite3
import random
import csv
```

```python
def to_csv(df, filename='output.csv'):
    df.to_csv(filename, index=False)
    print(f"Data written to {filename}")

def to_json(df, filename='output.json'):
    df.to_json(filename, orient='records', lines=True)
    print(f"Data written to {filename}")

def to_sql(df, db_name='output.db', table_name='people'):
    conn = sqlite3.connect(db_name)
    df.to_sql(table_name, conn, if_exists='replace', index=False)
    conn.close()
    print(f"Data written to {db_name} in table {table_name}")
```
