```python
from pymongo import MongoClient
import pprint

connection_string = "mongodb+srv://username:password@clustername.mongodb.net/test?retryWrites=true&w=majority"

client = MongoClient(connection_string)

db = client['sample_mflix']
collection = db['movies']
```

```python
action_movie = collection.find_one({"genres": "Action"})
pprint.pprint(action_movie)

movies_after_2000 = collection.find({"year": {"$gt": 2000}}).limit(5)
pprint.pprint(movies_after_2000)

high_rated_movies = collection.find({"imdb.rating": {"$gt": 8.5}}).limit(5)
pprint.pprint(high_rated_movies)

action_adventure_movies = collection.find({"genres": {"$all": ["Action",
"Adventure"]}}).limit(5)
pprint.pprint(action_adventure_movies)
```

```python
sorted_comedy_movies = collection.find({"genres": "Comedy"}).sort("imdb.rating", -1).limit(5)
for movie in sorted_comedy_movies:
    pprint.pprint(movie)

sorted_drama_movies = collection.find({"genres": "Drama"}).sort("year", 1).limit(5)
for movie in sorted_drama_movies:
    pprint.pprint(movie)
```

```python
avg_rating_by_genre = collection.aggregate([
    {"$unwind": "$genres"},
    {"$group": {"_id": "$genres", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"avg_rating": -1}},
    {"$limit": 5}
])
for genre in avg_rating_by_genre:
    pprint.pprint(genre)

top_directors = collection.aggregate([
    {"$group": {"_id": "$director", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}},
    {"$limit": 5}
])
for director in top_directors:
    pprint.pprint(director)

total_movies_per_year = collection.aggregate([
    {"$group": {"_id": "$year", "count": {"$sum": 1}}},
    {"$sort": {"_id": 1}}
])
for movies in total_movies_per_year:
    pprint.pprint(movies)
```

```python
collection.update_one({"title": "The Godfather"},
                      {"$set": {"imdb.rating": 9.5}})

collection.update_many({"genres": "Horror", "imbd": {"$exists": False}},
                       {"$set": {"imbd.rating": 6.0}})

collection.delete_many({"year": {"$lt": 1950}})
```

```python
high_rated_action_movies = collection.find({"genres": "Action", "imdb.rating": {"$gt": 8}}).sort("year", -1)
for movie in high_rated_action_movies:
    pprint.pprint(movie)

chris_nolan_movies = collection.find({"director": "Christopher Nolan", "imdb.rating": {"$gt": 8}}).sort("imdb.rating", -1).limit(3)
for movie in chris_nolan_movies:
    pprint.pprint(movie)
```
